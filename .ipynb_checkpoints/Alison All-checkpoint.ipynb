```python
import pandas as pd
import numpy as np
import glob
import re
from urllib import request as req
import json
import time

reportnames = []
reportserial = []
isMgr = []
locality_= []
identity_=[]

start = time.time()

request = req.urlopen("https://unified-profile-api.us-south-k8s.intranet.ibm.com/v3/profiles/681658897/teamResolved")
datos = request.read().decode()
dictdatos = json.loads(datos)
reports = dictdatos['content']['functional']['reports']
for r in reports:
    #method setdefault(key,[]) sirve para poner una primera key con una lista vacia, si no, tira error de que no existe la key
    reportnames.append(r['nameDisplay'])
    reportserial.append(r['uid'])
    isMgr.append(r['isManager'])
    locality_.append(r['address']['business']['locality'])
    identity_.append(r['preferredIdentity'])
    
#reportlist = devnames['Alison Yara']

df_Yara = pd.DataFrame(reportnames, columns=['Reports'])
df_Yara['Serials'] = reportserial
df_Yara['isManager'] = isMgr
df_Yara['Location'] = locality_
df_Yara['Email'] = identity_
df_Yara['Manager'] = "Alison Yara"


def reportFunc(dataframe):
    reportnames_2 = []
    reportserial_2 = []
    isMgr_2 = []
    locality = []
    identity = []
    mgr_ = []

    for i in range(dataframe['Reports'].size):
        request = req.urlopen("https://unified-profile-api.us-south-k8s.intranet.ibm.com/v3/profiles/"+dataframe['Serials'][i]+"/teamResolved")
        datos = request.read().decode()
        dictdatos = json.loads(datos)
        try:
            reports = dictdatos['content']['functional']['reports']
            for r in reports:
                #method setdefault(key,[]) sirve para poner una primera key con una lista vacia, si no, tira error de que no existe la key
                reportnames_2.append(r['nameDisplay'])
                reportserial_2.append(r['uid'])
                try:
                    email = r['preferredIdentity']
                    identity.append(r['preferredIdentity'])
                except:
                    identity.append('Contractor or LoA')
                isMgr_2.append(r['isManager'])
                try:
                    location = r['address']['business']['locality']
                    locality.append(r['address']['business']['locality'])
                except:
                    locality.append('Contractor')
                mgr_.append(dataframe['Reports'][i])
        except:
            pass
    return reportnames_2, reportserial_2, isMgr_2, mgr_, locality, identity

r1, rs1, ism1, mgr1, locality1, identity1 = reportFunc(df_Yara)

df_Reports = pd.DataFrame(r1, columns=['Reports'])
df_Reports['Serials'] = rs1
df_Reports['isManager'] = ism1
df_Reports['Manager'] = mgr1
df_Reports['Location'] = locality1
df_Reports['Email'] = identity1

r2, rs2, ism2, mgr2, locality2, identity2 = reportFunc(df_Reports)

df_Reports_2 = pd.DataFrame(r2, columns=['Reports'])
df_Reports_2['Serials'] = rs2
df_Reports_2['isManager'] = ism2
df_Reports_2['Manager'] = mgr2
df_Reports_2['Location'] = locality2
df_Reports_2['Email'] = identity2

r3, rs3, ism3, mgr3, locality3, identity3 = reportFunc(df_Reports_2)

df_Reports_3 = pd.DataFrame(r3, columns=['Reports'])
df_Reports_3['Serials'] = rs3
df_Reports_3['isManager'] = ism3
df_Reports_3['Manager'] = mgr3
df_Reports_3['Location'] = locality3
df_Reports_3['Email'] = identity3

r4, rs4, ism4, mgr4, locality4, identity4 = reportFunc(df_Reports_3)

df_Reports_4 = pd.DataFrame(r4, columns=['Reports'])
df_Reports_4['Serials'] = rs4
df_Reports_4['isManager'] = ism4
df_Reports_4['Manager'] = mgr4
df_Reports_4['Location'] = locality4
df_Reports_4['Email'] = identity4

r5, rs5, ism5, mgr5, locality5, identity5 = reportFunc(df_Reports_4)

df_Reports_5 = pd.DataFrame(r5, columns=['Reports'])
df_Reports_5['Serials'] = rs5
df_Reports_5['isManager'] = ism5
df_Reports_5['Manager'] = mgr5
df_Reports_5['Location'] = locality5
df_Reports_5['Email'] = identity5
    
df_all = df_Yara.append(df_Reports)
df_all = df_all.append(df_Reports_2)
df_all = df_all.append(df_Reports_3)
df_all = df_all.append(df_Reports_4)
df_all = df_all.append(df_Reports_5)


df_all.to_excel(r"C:\Users\016434613\Desktop\All_report.xlsx", index=False)

end = time.time()
mins = (end-start)/60.00
print("Report created in: "+str(mins)+" minutes")
```
